```python
# Function to convert a string to int or float, returns type info
def convert_num(s):
    try:
        return int(s), 'int'
    except:
        try:
            return float(s), 'float'
        except:
            return None, 'invalid'

# Arithmetic operation functions
def add(a, b):
    return a + b

def minus(a, b):
    return a - b

def product(a, b):
    return a * b

def divide(a, b):
    return a / b

def power(a):
    return a ** a  # Raises a to the power of a

# Display menu
print("Choose an operation:")
print("a = Addition")
print("b = Subtraction")
print("c = Product")
print("d = Division")
print("e = Power (a^a)")

# Take user choice
user = input("Enter your choice (a/b/c/d/e): ")

# Process only valid operations
if user in ['a', 'b', 'c', 'd', 'e']:
    a_str = input("Enter first number: ")
    b_str = input("Enter second number: ") if user != 'e' else '0'  # Skip second input for power

    a, a_type = convert_num(a_str)
    b, b_type = convert_num(b_str)

    # Validate inputs
    if a is None or (user != 'e' and b is None):
        print("❌ Please enter valid numbers only!")
    else:
        # Perform selected operation
        if user == 'a':
            res = add(a, b)
        elif user == 'b':
            res = minus(a, b)
        elif user == 'c':
            res = product(a, b)
        elif user == 'd':
            if b == 0:
                print("❌ Cannot divide by zero.")
                exit()
            res = divide(a, b)
        elif user == 'e':
            res = power(a)

        # Format result: show integer if both inputs were int and result is whole number
        if (a_type == 'int' and b_type == 'int' and user not in ['d', 'e']):
            res = int(res)
        else:
            res = float(res)

        print(f"✅ Result: {res}")
else:
    print("❌ Invalid operation selected.")
```
